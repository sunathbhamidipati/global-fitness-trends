```python
import pandas as pd
import glob

all_data = []

# Our clean, standardized column names (Date + your 5 terms)
standard_columns = ['Date', 'Gym', 'Cycling', 'Running', 'Yoga', 'Aerobics']

print("Starting to stitch and clean files...")

for file in glob.glob("time_series_*.csv"):
    # Extract the country code from the file name
    country_code = file.split('_')[2] 
    
    try:
        # skiprows=2 skips the weird Google metadata at the very top
        df = pd.read_csv(file, skiprows=1) 
        
        # Check to make sure the file has exactly 6 columns before renaming
        if len(df.columns) == 6:
            # Force the column names to be identical across all files
            df.columns = standard_columns
            
            # Add our primary key
            df['Country'] = country_code
            all_data.append(df)
            print(f"Processed cleanly: {country_code}")
        else:
            print(f"Skipping {country_code}: File had {len(df.columns)} columns instead of 6.")
            
    except Exception as e:
        print(f"Error processing {file}: {e}")

if all_data:
    master_trends = pd.concat(all_data, ignore_index=True)
    
    # Save the final, clean master file
    master_trends.to_csv("master_global_fitness_trends.csv", index=False)
    print(f"\nSuccess! Merged {len(all_data)} files into a perfectly aligned dataset.")
else:
    print("No valid files were processed.")
```
